```python
import pandas as pd
import altair as alt
from altmol import encode_molecules, mol_plot
```

```python
df = pd.read_csv("aizythfinder_top_templates.csv")
```

```python
df_aizy = encode_molecules(df, smiles_col_name='retro_template', img_format='svg', img_width=300, img_height=100)
```

```python
df_aizy = df_aizy.sort_values(by=["library_occurence"], ascending=False, ignore_index=True)
df_aizy["index"] = df_aizy.index
```

```python
bar_chart = mol_plot(
    df_aizy,
    x_axis=alt.X("index", scale=alt.Scale(domain=[0, 100])),
    y_axis=alt.Y("library_occurence"),
    selector=False,
    interactive=False,
    title="Aizynthfinder templates",
    chart_type="bar",
    mark_fill=True,
    mark_size=5,
)
```
